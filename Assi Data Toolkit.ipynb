```python
# Difference between np.array, np.asarray and np.asany array
```

```python
# np.array: creates a new array and copies data
a=np.array([1, 2, 3])
b=np.array(a)
b
```

```python
# np.asarray: converts inputs to arrays without copying if the input is already an array.
a=np.array([1, 2, 3])
b=np.asarray(a)  # b copies the array data of a
print("b array",b)
c=np.asarray([4, 5, 6]) # creates new array
print("c array",c)
```

```python
# np.asanyarray: same as np.asarray but retains array subclasses
a=np.matrix([[1,2],[3,4]])
b=np.asanyarray(a)           # b is a matrix as a
print("b array",b)
print("Data Type of B:",type(b))
c=np.asanyarray([5,6,7])     # creates new array
print("c array",c)
```

```python
# Difference between deep copy and shallow copy.
```

```python
# Deep copy: creates a new object and recursively copies all nested objects.Changes to the original do not affect the copy.
import copy
ori=[[1,2,3],[4,5,6]]
deep=copy.deepcopy(ori)
ori[0][0] = 'X'
print("Original:",ori)
print("Deep Copied:",deep)
```

```python
# Shallow Copy: creates a new object but references nested objects. Changes to nested objects affect both the original and the copy.
ori=[[1,2,3],[4,5,6]]
shal=copy.copy(ori)
ori[0][0] = 'X'
print("Original:",ori)
print("Shallow Copied:",shal)
```

```python
# 14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
# b) Label the x-axis as 'Date'.
# c) Set the title of the plot as 'Temperature and Humidity Over Time'.
```

```python
dat=pd.date_range(start='2024-01-01',periods=10,freq='D')
temp=np.random.uniform(10,41,size=10)
humi=np.random.uniform(0,100,size=10)
df=pd.DataFrame({'Date':dat,'Temperature':temp,'Humidity':humi})
fig,ax1=plt.subplots()
ax1.plot(df['Date'],df['Temperature'],color='red',label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature')
ax2=ax1.twinx()
ax2.plot(df['Date'],df['Humidity'],color='blue',label='Humidity')
ax2.set_ylabel('Humidity')
plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
# 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.
# b) Overlay a line plot repesenting the normal distribution's probability density function (PDF).
# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
# d) Set the title of the plot as 'Histogram with PDF Overlay'.
```

```python
# Ans
arr=np.random.normal(size=1000)
plt.hist(arr,bins=30,density=True)
xmin,xmax=plt.xlim()
x=np.linspace(xmin,xmax,100)
pdf = (1 /(np.sqrt(2 * np.pi))) * np.exp(-0.5 * (x) ** 2)
plt.plot(x, pdf,label='PDF')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.legend()
plt.show()
```

```python
# 17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
```

```python
# Ans
arr1=np.random.randn(100)
arr2=np.random.randn(100)
df=pd.DataFrame({'x':arr1,'y':arr2})
def quad(a):
    if a['x']>=0 and a['y']>=0:
        return 'Quadrant 1'
    elif a['x']<0 and a['y']>=0:
        return 'Quadrant 2'
    elif a['x']<0 and a['y']<0:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'
df['Quadrant'] = df.apply(quad, axis=1)
plt.figure()
scatter = sns.relplot(df,x='x',y='y',hue='Quadrant',palette='Set1',s=100)
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
# 18. With Bo+eh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
```

```python
from bokeh.plotting import figure,show
x=np.linspace(-10,10,100)
y=np.sin(x)
p=figure(title='Sine Wave Function',x_axis_label='X',y_axis_label='sin(X)')
p.line(x,y,legend_label='sin(x)')
show(p)
```

```python
# 19. 8 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.
```

```python
from bokeh.models import ColumnDataSource, HoverTool
cat=[f'Category {i}' for i in range(1, 11)]
val=np.random.randint(1,100,size=len(cat))
df=pd.DataFrame({'Category':cat,'Value':val})
sor=ColumnDataSource(df)
p=figure(x_range=df['Category'],title='Random Categorical Bar Chart',
           x_axis_label='Categories',y_axis_label='Values',
           tools='hover',tooltips="@Category: @Value")
p.vbar(x='Category',top='Value',source=sor,
       line_color='white',fill_color='blue',legend_field='Category')
p.legend.orientation="horizontal"
p.legend.location="top_center"
show(p)
```

```python
# 20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'.
```

```python
import plotly.graph_objs as go
x=np.linspace(0,10,100)
y=np.random.rand(100)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines'))
fig.update_layout(title='Simple Line Plot',xaxis_title='X-axis',yaxis_title='Y-axis')
fig.show()
```
